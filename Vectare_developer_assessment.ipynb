**Problem 1:**
If we list all the natural numbers below 10 that are multiples of 3 or 5, we
get 3, 5, 6 and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000.

```python
def sum_of_3s_and_5s():
  sum = 0
  for i in range(1000):
      if i % 3 == 0 or i % 5 == 0:
          sum += i
  return sum
print(sum_of_3s_and_5s())
```

**Prblem3:** A palindromic number reads the same both ways. The largest palindrome
made from the product of two 2-digit numbers is 9009 = 91 × 99.
Find the largest palindrome made from the product of two 3-digit
numbers.

```python
def is_palindromic(n: int) -> bool:
  return str(n) == str(n)[::-1]

def largetest_number() -> int:
  max = 0
  for i in range(999, 99, -1):
    for j in range(i, 99, -1):
      num = i * j
      if num > max and is_palindromic(num):
        max = num
  return max

print(largetest_number())
```

**Problem 4:**2520 is the smallest number that can be divided by each of the numbers
from 1 to 10 without any remainder. What is the smallest positive number that is evenly divisible by all of the numbers from 1 to 20?

```python
#lcm(a, b) = a * b / gcd(a, b)
import math
def smallest_lcm(n) -> int:
  lcm = 1
  for i in range(1, n + 1):
    lcm = lcm * i // math.gcd(lcm, i)
  return lcm

print(smallest_lcm(20))
```

Problem 5: By listing the first six prime numbers: 2, 3, 5, 7, 11, and 13, we can see that the 6th prime is 13.
What is the 10,001st prime number?

```python
def is_prime(n: int) -> bool:
  if n < 2:
    return False
  if n == 2:
    return True
  if n % 2 == 0:
    return False
  for i in range(3, int(n ** 0.5) + 1, 2):
    if n % i == 0:
      return False
  return True

def prime_number(n: int) -> int:
  if n == 1:
    return 2

  count = 1
  num = 1
  while count < n:
    num += 2
    if is_prime(num):
      count += 1
  return num

print(prime_number(10001))
```

**Promblem 6:** Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right
corner. How many such routes are there through a 20×20 grid?

```python
import math
#1x1: 2
#2x2: 6
#n moves to right, n moves to down, total 2n moves
#path = choose right/down n moves in 2n moves
#2n!/ n! * n!
def count_path(n: int) -> int:
  return math.comb(2 * n, n)

print(count_path(20))
```

**Problem 7:** By starting at the top of the triangle below and moving to adjacent numbers on the row below, the maximum total from top to bottom is 23.
3
7 4
2 4 6
8 5 9 3
That is, 3 + 7 + 4 + 9 = 23.
Find the maximum total from top to bottom in triangle.txt (attached to
the email), a 15K text file containing a triangle with one-hundred rows.

```python
#Bottom-up
#dp
def maximum_total(triangle: list[list[int]]) -> int:
  for row in range(len(triangle)-2, -1, -1):
    for col in range(len(triangle[row])):
      triangle[row][col] += max(triangle[row+1][col], triangle[row+1][col+1])
  return triangle[0][0]

def read_file(filename: str) -> list[list[int]]:
  with open(filename, 'r') as f:
    return [[int(num) for num in line.split()] for line in f]

triangle = read_file('triangle.txt')
print(maximum_total(triangle))
```
